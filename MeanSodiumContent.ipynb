```python
# https://youtu.be/rlrMOphKy4w

import numpy as np, pandas as pd, matplotlib.pyplot as plt, seaborn as sns, scipy.stats as stats, seaborn as sns
```

```python
df = pd.read_csv('./data/FF_sodium.csv')
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


# Bootstrapping
n_iterations = 5000
means = []

for _ in range(n_iterations):
    sample = df.sample(frac=1, replace=True)
    sample_mean = sample['Sodium(mg)'].mean()
    means.append(sample_mean)

# Confidence interval
alpha = 0.05
lower_bound = np.percentile(means, alpha/2*100)
upper_bound = np.percentile(means, (1-alpha/2)*100)

print(f'95% Confidence Interval: [{lower_bound:.2f}, {upper_bound:.2f}]')
print(f'Bootstrap Samples Mean: {np.mean(means):.2f}\nSample Mean: {np.mean(df['Sodium(mg)']):.2f}')

# Histogram
counts, bins, patches = plt.hist(means, bins=30, edgecolor='k', alpha=0.7)

# Color the bars
for count, patch in zip(counts, patches):
    if patch.get_x() + patch.get_width() < lower_bound or patch.get_x() > upper_bound:
        patch.set_facecolor('red')

plt.title('Bootstrap Sample Means')
plt.xlabel('Mean Sodium (mg)')
plt.ylabel('Frequency')
plt.axvline(np.mean(means), color='green', linestyle='dotted', linewidth=3)
plt.axvline(lower_bound, color='r', linestyle='dashed', linewidth=1)
plt.axvline(upper_bound, color='r', linestyle='dashed', linewidth=1)
plt.show()
```

This code will:

1. Create a DataFrame from the provided sodium data.
2. Perform bootstrapping by repeatedly sampling with replacement and calculating the sample mean.
3. Calculate the 90% confidence interval for the sample mean.
4. Plot a histogram of the bootstrap sample means, with the confidence interval bounds marked
5. Color the bars red if they are greater than the upper bound or less than the lower bound.
